```python
import os
import pandas as pd

# User inputs
experiment = "PlasmaTreated_08-01-2025"
  # Define the strain name 
username = "spack"  # Your username for your computer

# Base directory and file pattern
#C:\Users\spack\OneDrive\Desktop\ImageJ\Raw_Images\Dispersal Experiment
base_dir = f'/Users/{username}/OneDrive/Desktop/ImageJ/Raw_Images/Dispersal Experiment/'
#f'/Users/{username}/OneDrive - Worcester Polytechnic Institute (wpi.edu)/Documents/Data/3D_Cluster Data/{experiment}'
file_pattern = experiment + "_W{}S{}_r_max=1.5.xlsx"

# Columns to append
columns_to_append = ["N_cl", 'Radius of gyration', "Diameter (Max Euclidean Distance)", "Shape anisotropy index", "Asphericity index"]

# Initialize an empty DataFrame to store the merged data
merged_data = pd.DataFrame()

# Iterate through different W and S values
for w in range(1, 5):  # Assuming W1, W2, W3, W4
    for s in range(1, 6):  # Assuming S1, S2, ..., S10
        # Construct the file path
        file_path = os.path.join(base_dir, file_pattern.format(w, s))
        
        # Check if the file exists
        if os.path.exists(file_path):
            try:
                # Read the Excel file
                df = pd.read_excel(file_path, engine='openpyxl')
                
                # Append the required columns to the merged DataFrame
                merged_data = merged_data.append(df[columns_to_append], ignore_index=True)
                
                print(f"Successfully processed file: {file_path}")
            except Exception as e:
                print(f"Error processing {file_path}: {e}")
        else:
            print(f"File does not exist: {file_path}")

# Save the merged data to a new Excel file
output_file = os.path.join(base_dir, experiment + "_SizeMorphology_Data.xlsx")
merged_data.to_excel(output_file, index=False)

print(f"Merged data saved to {output_file}")
```
